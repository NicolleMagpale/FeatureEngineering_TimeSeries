```python
#SIMPLE DATA READING
from pandas import Series
from pandas import DataFrame
from pandas import concat
import pandas as pd
import os
PATH = 'merged_mmda_weatherforyou_manila_2015_bystation\\'
```

```python

def generateFeatures(variable):

    # Generate Shifts
    series[variable + '_t-10'] = series[variable].shift(10)
    series[variable + '_t-9'] = series[variable].shift(9)
    series[variable + '_t-8'] = series[variable].shift(8)
    series[variable + '_t-7'] = series[variable].shift(7)
    series[variable + '_t-6'] = series[variable].shift(6)
    series[variable + '_t-5'] = series[variable].shift(5)
    series[variable + '_t-4'] = series[variable].shift(4)
    series[variable + '_t-3'] = series[variable].shift(3)
    series[variable + '_t-2'] = series[variable].shift(2)
    series[variable + '_t-1'] = series[variable].shift(1)
    series[variable + '_t'] = series[variable]


    #Expanding Part
    expandingWindow = series[variable].expanding()

    expandingWidth = 5
    expandingMeans = expandingWindow.mean()
    expandingMins = expandingWindow.min()
    expandingMaxs = expandingWindow.max()
    expandingSum = expandingWindow.sum()
    series[variable + '_Emean'] = expandingMeans
    series[variable + '_Emin'] = expandingMins
    series[variable + '_Emax'] = expandingMaxs
    series[variable + '_Esum'] = expandingSum

    # Rolling Window Features window 3 = 30mins

    rolllingWidth = 3
    rolllingWindow = series[variable + '_t-2'].rolling(window=rolllingWidth)
    rollingMeans = rolllingWindow.mean()
    rollingMins = rolllingWindow.min()
    rollingMaxs = rolllingWindow.max()
    rollingMaxs = rolllingWindow.sum()
    series[variable + '_Rmean (window = 3)'] = rollingMeans
    series[variable + '_Rmin  (window = 3)'] = rollingMins
    series[variable + '_Rmax  (window = 3)'] = rollingMaxs

    # Rolling Window Features window 5 = 1hr

    rolllingWidth = 5
    rolllingWindow = series[variable + '_t-4'].rolling(window=rolllingWidth)
    rollingMeans = rolllingWindow.mean()
    rollingMins = rolllingWindow.min()
    rollingMaxs = rolllingWindow.max()
    rollingMaxs = rolllingWindow.sum()
    series[variable + '_Rmean (window = 5)'] = rollingMeans
    series[variable + '_Rmin  (window = 5)'] = rollingMins
    series[variable + '_Rmax  (window = 5)'] = rollingMaxs

    # Rolling Window Features window 7 = 1hr 30mins

    rolllingWidth = 7
    rolllingWindow = series[variable + '_t-6'].rolling(window=rolllingWidth)
    rollingMeans = rolllingWindow.mean()
    rollingMins = rolllingWindow.min()
    rollingMaxs = rolllingWindow.max()
    rollingMaxs = rolllingWindow.sum()
    series[variable + '_Rmean (window = 7)'] = rollingMeans
    series[variable + '_Rmin  (window = 7)'] = rollingMins
    series[variable + '_Rmax  (window = 7)'] = rollingMaxs

    
    # Rolling Window Features window 9 = 2hrs

    rolllingWidth = 9
    rolllingWindow = series[variable + '_t-8'].rolling(window=rolllingWidth)
    rollingMeans = rolllingWindow.mean()
    rollingMins = rolllingWindow.min()
    rollingMaxs = rolllingWindow.max()
    rollingMaxs = rolllingWindow.sum()
    series[variable + '_Rmean (window = 9)'] = rollingMeans
    series[variable + '_Rmin  (window = 9)'] = rollingMins
    series[variable + '_Rmax  (window = 9)'] = rollingMaxs
```

```python
for filename in os.listdir(PATH):
    print(filename)
    series = pd.read_csv(PATH + filename, skipinitialspace = True)
    for v in range(2,4):
        variable = series.columns.values[v]
        generateFeatures(variable)
    for v in range(6,12):
        variable = series.columns.values[v]
        generateFeatures(variable)
    series.to_csv("Engineered\Engineered_" + filename, encoding='utf-8', index=False)


```
